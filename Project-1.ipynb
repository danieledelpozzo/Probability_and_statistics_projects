## FUNCTION DEFINITION

lcg

```python
def lcg(seed, a, c, m, n, rescale):
    
    random_numbers = np.empty(n)
    
    # Set the sequence to the initial value
    x = seed
    random_numbers[0] = x
    
    for i in range(1, n):
        x = (a * x + c) % m
        random_numbers[i] = x
    
    if rescale:
        random_numbers = random_numbers / m
    
    return random_numbers
```

msm

```python
def msm(seed, n):
    # Initialize an empty list to store the random numbers
    sequence = []
    
    # Generate 'n' pseudo-random numbers
    for _ in range(n):

        # Square the seed and convert it to a string
        # Pad with zeros to ensure it has at least 8 digits
        # Extract the middle 4 digits from the squared result
        seed = int(str(seed ** 2).zfill(8)[2:6])
        
        # Append the newly generated number to the sequence
        sequence.append(seed)
        
    # Return the complete list of generated numbers
    return sequence
```
